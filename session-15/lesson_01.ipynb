```python
x=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday','Sunday']

z=np.random.rand(3,4)
y=np.random.rand(7)*10
z=y-2
w=list(np.random.randint(3,20,7))
plt.subplot(2,1,1)
plt.plot(x,y, c='green', marker='o', markersize=10)
plt.subplot(2,1,2)
plt.plot(x,z, c='brown', marker='*', markersize=10)

plt.title('Test Plot', color='red', fontname='Tahoma', fontsize=20)

plt.xlabel('values', color='red', fontname='Tahoma', fontsize=20)
plt.ylabel('sin', color='red', fontname='Tahoma', fontsize=20)

plt.annotate(xy=['Saturday',2], s='annotate')

plt.legend(['y','z'], loc=0)

plt.style.use('grayscale')

# set the spacing between subplots
plt.subplots_adjust(left=0.1,
					bottom=0.1, 
					right=0.9, 
					top=0.9, 
					wspace=0.4, 
					hspace=0.4)
plt.show()

plt.savefig('test.pdf', format='pdf')

plt.show()
```

```python
# making subplots with constrained_layout=True
x=y
fig, ax = plt.subplots(2, 2, 
					constrained_layout = True)

# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 1].plot(x, x*2)
ax[1, 0].plot(x, x*x)
ax[1, 1].plot(x, x*x*x)
ax[0,0].set_title('Test', color='red')
ax[0,1].set_title('dlkjlsakdj')
ax[0,0].set_xlabel('x')

ax[0,1].legend(['x','y'])

fig.suptitle('Main title')
plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d

from matplotlib import style

style.use('ggplot')

fig=plt.figure()

ax1=fig.add_subplot(111, projection='3d')

x3=np.random.randint(1,10,9)
y3=np.random.randint(3,12,9)
z3=np.random.randint(18,27,9)

dx=np.ones(9)
dy=np.ones(9)
dz=[1,2,3,4,5,6,7,8,9]

ax1.bar3d(x3,y3,z3,dx,dy,dz,color='red')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')

fig.suptitle('Categorical')
plt.show()
```
